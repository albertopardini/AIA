```python
import pandas as pd

# Ruta al archivo CSV en Colab
rutaArchivo = '/content/calidad_aire_datos_mes.csv'

calidadAireDatosMes = pd.read_csv(rutaArchivo, thousands='.', decimal=',',sep=';')

# Mostrar las primeras filas del conjunto de datos
calidadAireDatosMes
```

Eliminaci√≥n de nulos, quitando las filas donde al menos una de las columnas h est√°n vac√≠as.

```python
# Especificar las columnas desde 'h01' hasta 'h24'
columns_to_check = [f'h{i:02d}' for i in range(1, 25)]

# Borrar las filas donde cualquiera de las columnas especificadas tenga un valor nulo
calidadAireDatosMes = calidadAireDatosMes.dropna(subset=columns_to_check)
calidadAireDatosMes
```

Eliminaci√≥n de columnas que no aportan al an√°lisis que estamos encarando como **provincia, punto de muestreo y v01** hasta **v24**.

```python
# Especificar las columnas a eliminar
columns_to_drop = ['provincia', 'punto_muestreo'] + [f'v{i:02d}' for i in range(1, 25)]
# Eliminar las columnas especificadas
calidadAireDatosMes = calidadAireDatosMes.drop(columns=columns_to_drop)
```

```python
calidadAireDatosMes
```

```python
# Listar dado un n√∫mero de 'magnitud', las filas que cumplen con la condici√≥n.
# Por ejemplo 'magnitud'=6 se trata de Mon√≥xido de Carbono.
calidadAireDatosMes[calidadAireDatosMes['magnitud']==6]
```

```python
# Listar dada un n√∫mero de 'magnitud' y un n√∫mero de 'municipio' las filas que cumplen con la condici√≥n.
# Por ejemplo 'magnitud'=6 se trata de Mon√≥xido de Carbono y 'municipio'=5 ALCALA DE HENARES.
calidadAireDatosMes[(calidadAireDatosMes['magnitud']==6) & (calidadAireDatosMes['municipio']==5)]
```

```python
# Listar dada un n√∫mero de 'magnitud',un n√∫mero de 'municipio' y un 'dia' del mes;las filas que cumplen con la condici√≥n.
# Por ejemplo 'magnitud'=6 se trata de Mon√≥xido de Carbono,'municipio'=5 ALCALA DE HENARES y 'dia'=1
calidadAireDatosMes[(calidadAireDatosMes['magnitud']==6) & (calidadAireDatosMes['municipio']==5) & (calidadAireDatosMes['dia']==1)]
```

Incorporamos al DataFrame las columnas **min_dia, max_dia y prom_day**  filtrando todas las columnas que empiezan con **h** y luego utilizando **min, max y mean** con **axis=1** para que calcule m√≠nimo, m√°ximo y promedio por filas. ‚ùó

```python
# Seleccionar las columnas desde 'h0' hasta 'h24'
columnasHoras = calidadAireDatosMes.filter(like='h')
minimoPorFila = columnasHoras.min(axis=1).round(2)
maximoPorFila = columnasHoras.max(axis=1).round(2)
promedioPorFila = columnasHoras.mean(axis=1).round(2)

# Agrega al dataframe la columna 'prom_dia' y 'max_dia'
calidadAireDatosMes ['min_dia'] = minimoPorFila
calidadAireDatosMes ['max_dia'] = maximoPorFila
calidadAireDatosMes ['prom_dia'] = promedioPorFila
```

Agregaremos al DataFrame una columna con la descripci√≥n de la magnitud.  
**magnitud_descrip**

```python
# Definir un diccionario para mapear los c√≥digos de magnitud a sus descripciones
magnitud_descripcion = {
    1:'Di√≥xido de azufre',
    6:'Mon√≥xido de carbono',
    7:'Mon√≥xido de nitr√≥geno',
    8:'Di√≥xido de nitr√≥geno',
    9:'Part√≠culas en suspensi√≥n < PM2.5',
    10:'Part√≠culas en suspensi√≥n< PM10',
    11:'Part√≠culas en suspensi√≥n< PM1',
    12:'√ìxidos de nitr√≥geno',
    14:'Ozono',
    20:'Tolueno',
    22:'Black Carbon',
    30:'Benceno',
    42:'Hidrocarburos totales',
    44:'Hidrocarburos no met√°nicos',
    431:'MetaParaXileno'
}

# Crear la columna 'magnitud_descrip' utilizando el diccionario
calidadAireDatosMes['magnitud_descrip'] = calidadAireDatosMes['magnitud'].map(magnitud_descripcion)

# Mostrar las primeras filas del DataFrame actualizado
calidadAireDatosMes.head()
```

Agregaremos al DataFrame una columna **nombre_estacion** para mapear el CODIGO MUNICIPIO con el NOMBRE ESTACION.

```python
# Definir un diccionario para mapear los c√≥digos de municipios a sus nombres de estaci√≥n.
nombre_estacion = {
    5:'ALCALA DE HENARES',
    6:'ALCOBENDAS',
    7:'ALCORCON',
    9:'ALGETE',
    13:'ARANJUEZ',
    14:'ARGANDA DEL REY',
    16:'ATAZAR, EL',
    45:'COLMENAR VIEJO',
    47:'COLLADO VILLALBA',
    49:'COSLADA',
    58:'FUENLABRADA',
    65:'GETAFE',
    67:'GUADALIX DE LA SIERRA',
    74:'LEGANES',
    80:'MAJADAHONDA',
    92:'MOSTOLES',
    102:'ORUSCO DE TAJU√ëA',
    106:'PARLA',
    115:'POZUELO DE ALARC√ìN',
    120:'PUERTO DE COTOS',
    123:'RIVAS-VACIAMADRID',
    127:'ROZAS, LAS',
    133:'SAN MART√çN DE VALDEIGLESIAS',
    134:'SAN SEBASTIAN DE LOS REYES',
    148:'TORREJ√ìN DE ARDOZ',
    161:'VALDEMORO',
    171:'VILLA DEL PRADO',
    180:'VILLAREJO DE SALVAN√âS'
}

# Crear la columna 'nombre_estacion' utilizando el diccionario
calidadAireDatosMes['nombre_estacion'] = calidadAireDatosMes['municipio'].map(nombre_estacion)

# Mostrar las primeras filas del DataFrame actualizado
calidadAireDatosMes.tail()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20
import pandas as pd
"""

GR√ÅFICO DE LINEAS CON DISPERSI√ìN (o l√≠neas con puntos).
En este gr√°fico:

Las l√≠neas representan la evoluci√≥n diaria de los valores m√°ximos (max_dia) para distintas categor√≠as (magnitud_descrip)
como ser "Ozono," "Tolueno" y "Mon√≥xido de carbono".

Los puntos dispersos (scatter plot o puntos de dispersi√≥n) se superponen en las posiciones de cada d√≠a (dia)
para se√±alar valores espec√≠ficos.

"""
# Para visualizar el gr√°fico en un notebook
output_notebook()

calidadAireDM = calidadAireDatosMes[
    (calidadAireDatosMes['magnitud_descrip'] == 'Ozono')|
    (calidadAireDatosMes['magnitud_descrip'] == 'Tolueno')|
    (calidadAireDatosMes['magnitud_descrip'] =='Mon√≥xido de carbono')
]

df = calidadAireDM[['dia', 'magnitud_descrip', 'max_dia']]

# Crear una figura
p = figure(x_axis_type="linear", title="MAXIMOS POR MAGNITUDES",
           x_axis_label="D√≠a", y_axis_label="Max Dia", height=400, width=700)

# Colores para cada categor√≠a de magnitud
colores = Category20[20]

# Graficar cada grupo de magnitud_descrip
for i, (magnitud, group) in enumerate(df.groupby('magnitud_descrip')):
    # Crear un ColumnDataSource para cada grupo
    source = ColumnDataSource(group)

    # Agregar una l√≠nea para cada categor√≠a de magnitud
    p.line(x='dia', y='max_dia', source=source, color=colores[i], legend_label=magnitud, line_width=2)
    p.scatter(x='dia', y='max_dia', source=source, color=colores[i], size=6)  # Usa scatter en lugar de circle

# Configuraciones adicionales
p.legend.title = "Magnitud Descrip"
p.legend.location = (100, 250)
p.legend.click_policy = "hide"  # Permite ocultar una l√≠nea al hacer clic en la leyenda

# Mostrar el gr√°fico
show(p)
```
