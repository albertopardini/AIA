###Available paramaters for the ```history()``` method are üìê

**period:** data period to download (Either Use period parameter or use start and end).    
Valid periods are: **1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max**
**interval:** data interval (intraday data cannot extend last 60 days).    
Valid intervals are: **1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo,
3mo**   
**start:** If not using period - Download start date string (YYYY-MM-DD) or datetime.   
**end:** If not using period - Download end date string (YYYY-MM-DD) or datetime.   
**prepost:** Include Pre and Post market data in results? (Default is False)   
**auto_adjust:** Adjust all OHLC automatically? (Default is True)   
**actions:** Download stock dividends and stock splits events? (Default is True)

````python
from datetime import datetime

# Timestamp Unix en segundos
#"'lastDividendDate': 1723680000,
timestamp=msft.info['lastDividendDate']

# Convertir a fecha legible
fecha = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
print(fecha)
````

````python
# get historical market data
msft.history(period="6mo")
````

````python
# show actions (dividends, splits)
msft.actions
````

````python
# show dividends
msft.dividends
````

````python
# show splits
msft.splits
````

###Mass download of market data

````python
#To access the closing price data for SPY, you should use: data['Close']['SPY'].
#If, however, you want to group data by Symbol, use:

import yfinance as yf
data = yf.download("SPY AAPL", start="2024-10-20", end="2024-10-23",
                   group_by="ticker")

# To access the closing price data for SPY, you should use: data['SPY']['Close'].
# The download() method accepts an additional parameter
# threads for faster completion when downloading a lot of symbols at once.

data.sort_values(by='Date', ascending=False)
````
